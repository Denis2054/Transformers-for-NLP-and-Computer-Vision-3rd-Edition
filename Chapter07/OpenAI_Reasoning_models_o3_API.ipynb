OpenAI's o3-mini model represents a significant advancement over the earlier o1 model, particularly in areas requiring complex reasoning, such as science, technology, engineering, and mathematics (STEM). Here's an overview of the key improvements:

**1. Enhanced Reasoning Capabilities:**
o3-mini is designed to simulate deeper reasoning processes, enabling it to handle complex, multi-step problems more effectively than o1. This is achieved through a technique known as "simulated reasoning," where the model internally deliberates before generating responses, leading to more accurate and coherent outputs.

**2. Improved Performance in STEM Domains:**
Evaluations have demonstrated that o3-mini outperforms o1 in various STEM-related tasks. For instance, in mathematics, o3-mini achieved higher accuracy on the American Invitational Mathematics Examination (AIME) compared to o1. Similarly, in coding benchmarks like Codeforces, o3-mini demonstrated superior performance, indicating its enhanced capability in programming tasks.

**3. Cost-Effectiveness and Efficiency:**
Despite its advanced capabilities, o3-mini is more cost-effective than o1. It delivers faster responses, with evaluations showing a 24% reduction in response time compared to o1-mini. This efficiency makes o3-mini a more practical choice for applications requiring quick and accurate reasoning.

**4. Flexibility with Reasoning Effort Levels:**
o3-mini offers adjustable reasoning effort settingslow, medium, and highallowing users to balance between speed and accuracy based on specific needs. This flexibility enables the model to "think harder" when tackling complex challenges or prioritize speed when latency is a concern.

**5. Advanced Safety Measures:**
OpenAI has implemented deliberative alignment techniques in o3-mini, training the model to reason about human-written safety specifications before responding to prompts. This approach enhances the model's ability to produce safe and reliable outputs, surpassing o1 in challenging safety evaluations.

In summary, o3-mini builds upon the foundation laid by o1, offering improved reasoning capabilities, performance, efficiency, flexibility, and safety. These enhancements make it a more effective tool for tackling complex problems in various domains.

## Running o3-mini through the chat completions endpoint

```python
prompt="Discuss the molecular mechanisms by which non-coding RNAs (ncRNAs) regulate gene expression in eukaryotic organisms. Include in your answer the roles of various classes of ncRNAs such as microRNAs (miRNAs), small interfering RNAs (siRNAs), piwi-interacting RNAs (piRNAs), and long non-coding RNAs (lncRNAs). Explain how these ncRNAs influence processes like chromatin remodeling, transcriptional control, RNA splicing, RNA editing, and mRNA stability. Provide examples of how dysregulation of ncRNAs is implicated in human diseases, such as cancer or neurodegenerative disorders, and discuss current therapeutic approaches targeting ncRNAs."
```

```python
import textwrap
print(textwrap.fill(str(prompt), 100))
```

```python
from openai import OpenAI
client = OpenAI()

response = client.chat.completions.create(
    model="o3-mini",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ]
)

print(response.choices[0].message.content)
```

```python
import markdown
# Extract the assistant's reply
markdown_content = response.choices[0].message.content

# Convert markdown to HTML
html_body = markdown.markdown(markdown_content, extensions=['fenced_code', 'tables'])

# Add CSS styling to enhance appearance
html_content = f"""
<style>
body {{
    font-family: Arial, sans-serif;
    line-height: 1.6;
    padding: 20px;
    color: #333;
}}
h1, h2, h3, h4, h5, h6 {{
    color: #2c3e50;
    margin-top: 24px;
}}
hr {{
    border: none;
    border-top: 1px solid #eee;
    margin: 24px 0;
}}
ul, ol {{
    margin-left: 20px;
}}
code {{
    background-color: #f9f9f9;
    padding: 2px 4px;
    font-size: 90%;
    color: #c7254e;
    border-radius: 4px;
}}
pre code {{
    background-color: #f9f9f9;
    padding: 10px;
    display: block;
}}
table {{
    border-collapse: collapse;
    width: 100%;
    margin-bottom: 24px;
}}
th, td {{
    text-align: left;
    padding: 8px;
    border: 1px solid #ddd;
}}
tr:nth-child(even) {{
    background-color: #f9f9f9;
}}
</style>
{html_body}
"""

# Display the HTML in the Colab notebook
from IPython.display import display, HTML
display(HTML(html_content))
```
